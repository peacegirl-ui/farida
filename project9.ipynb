#                              ===== Handle Missing Data =====



```python
#1)Displays rows with missing values
df[df.isnull().any(axis=1)]
```

```python
#fill missing values with mean
df["Age"]=df["Age"].fillna(df["Age"].mean())
df
```

```python
#drop rows with missing values
df_clean = df.dropna()
df_clean
```

```python
df=df_clean
```

```python
df.shape
```

```python
# Replace missing values with a specific values
df.loc[:, 'Status'] = df['Status'].fillna("Empty")
df.loc[:, 'unnamed1'] = df['unnamed1'].fillna("Empty")

```

```python
plt.figure(figsize=(8,5))

plt.scatter( data=df , y="Orders" , x="Age Group" , edgecolor="Black", alpha=0.6, marker="o")

plt.title("Generating Scatter plot")
plt.xlabel("Age Group")
plt.ylabel("Orders")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
