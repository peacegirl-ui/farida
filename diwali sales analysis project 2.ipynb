# Explotary Data analysis

```python
# plotting a bar chart for Gender and it's count
Fa=sns.countplot(x="Gender",data=df , hue="Gender")
sns.set(rc={"figure.figsize":(5,4)})
for bars in Fa.containers:
    Fa.bar_label(bars) # we pased this for labels
```

```python
sales_gender=df.groupby(["Gender"] , as_index=False)["Amount"].sum().sort_values(by="Amount", ascending =False)
sns.barplot(x="Gender",y="Amount", data=sales_gender , hue="Gender")
```

 From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

# Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group" ,y="Amount",data= sales_age , hue="Age Group")
```

*from above graph we can see that most of the buyers are females and are of age group betweet 26-35*

# state


```python
#total number of order from last 10 states
sales_states=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Orders", data=sales_states , hue="Orders")
```

```python
#total amount/sales for first 10 states
sales_states=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(18,8)})

sns.barplot(x="State",y="Amount",data=sales_states,hue="Amount")
```

*we cam see that most of the orders and total sales /amount  from UP ,Maharastra , Karnataka*

# Martial status
