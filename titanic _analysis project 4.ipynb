```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# TITANIC  REPORT ANALYSIS

```python
df=pd.read_csv("titanic_small_dataset (1).csv")
df
```

```python
df.shape #10 ROWS AND 12 COLUMN
```

```python
df= df.dropna(subset=["Survived" ,"Sex","Age"])
```

```python

df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df["Cabin"]=df["Cabin"].fillna("Unknown")
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df["Fare"]=df["Fare"].astype(int)
```

```python
df['Ticket_code'] = pd.factorize(df['Ticket'])[0]
df['Name_code'] = pd.factorize(df['Name'])[0]
df["Cabin_code"]=pd.factorize(df["Cabin"])[0]
```

```python
df.dtypes
```

```python
df.drop(["Cabin","Ticket","Name"], axis=1 , inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print(df['Survived'].value_counts())
```

```python
#survived based on gender(sex)
FT=sns.countplot(data=df,x="Sex", hue="Survived" ,palette=["pink","skyblue"]) #pink(0) - not survived , blue(1)- survived

for bars in FT.containers:
    FT.bar_label(bars)

plt.title("Survived by gender(sex)")
plt.show()
```

*from the above chart we can see that females are more survived as compared to males*

```python
#survived based on Pclass
FT=sns.countplot(data=df,x="Pclass", hue="Survived" ,palette=["red","green"]) #red(0) - not survived , green(1)- survived

for bars in FT.containers:
    FT.bar_label(bars)

plt.title("Survived by Pclass")
plt.show()
```

*this graph tells us survival chances are better in higher class*

```python
#age distribution
sns.set(rc={"figure.figsize":(7,5)})

sns.histplot(df[df["Survived"]==1]["Age"],color="green",label="Survived",kde=True) #survived
sns.histplot(df[df["Survived"]==0]["Age"],color="purple",label=" not Survived",kde=True) #not survived

plt.legend()
plt.title("Age vs survival")
plt.xlabel("Age")
plt.ylabel(" no of passengers")
plt.show()
```

1) *no of passengers below 15 years likely higher chance of survival (especially chrildens)*
2) *Most people who died were adults between 20–40 years old.*
3) *very few elderly people survived*

# Conclusion

*from the above graph we observed that females are more survived as compared to men*
*also survival chances are better in higher class*
